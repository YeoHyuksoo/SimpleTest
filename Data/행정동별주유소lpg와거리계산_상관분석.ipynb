```python
## import pandas as pd
from geopy.distance import geodesic

# 거리 계산 결과를 저장할 리스트
distances = []

# 각 동과 수소충전소 간 거리 계산
for i, emd_row in emd_data.iterrows():
    for j, suso_row in suso_data.iterrows():
        emd_coord = (emd_row['위도'], emd_row['경도'])
        suso_coord = (suso_row['위도'], suso_row['경도'])
        distance = geodesic(emd_coord, suso_coord).meters  # 거리(m)
        distances.append([emd_row['지도코드'], suso_row['수소충전소'], distance])

# 결과를 데이터프레임으로 변환
df_distances = pd.DataFrame(distances, columns=['지도코드', '수소충전소', '거리(m)'])
print(df_distances)
```

```python
## import pandas as pd
from geopy.distance import geodesic

# 거리 계산 결과를 저장할 리스트
distances = []

# 각 주유소와 민감시설들 간 거리 계산
for i, oil_row in oil_data.iterrows():
    for j, sens_row in sens_data.iterrows():
        oil_coord = (oil_row['위도'], oil_row['경도'])
        sens_coord = (sens_row['위도'], sens_row['경도'])
        distance = geodesic(oil_coord, sens_coord).meters  # 거리(m)
        distances.append([oil_data.iloc[i, 1], sens_data.iloc[j, 2], distance])

# 결과를 데이터프레임으로 변환
df_distances = pd.DataFrame(distances, columns=['oil_nm', 'sens_addr', '거리(m)'])
print(df_distances)
```

```python
# 1. 행정동별로 주유소 간 거리의 최솟값
min_distance = df_distances.groupby('지도코드')['거리(m)'].min().reset_index()
min_distance.columns = ['지도코드', '거리_최솟값']
min_distance
```

```python
# 2. 행정동별로 가장 작은 거리 2개의 평균
top2_mean_distance = (
    df_distances.groupby('지도코드')['거리(m)']
    .apply(lambda x: x.nsmallest(2).mean())
    .reset_index()
)
top2_mean_distance.columns = ['지도코드', '최소거리2개평균']
```

```python
# 결과를 하나의 데이터프레임으로 합치기
result = min_distance.merge(top2_mean_distance, on='지도코드')
print(result)
```

```python

# 1. 행정동별로 주유소 간 거리의 최솟값
min_distance = df_distances.groupby('행정동_index')['거리(m)'].min().reset_index()
min_distance.columns = ['행정동_index', '최솟값']

# 2. 행정동별로 가장 작은 거리 3개의 평균
top3_mean_distance = (
    df_distances.groupby('행정동_index')['거리(m)']
    .apply(lambda x: x.nsmallest(3).mean())
    .reset_index()
)
top3_mean_distance.columns = ['행정동_index', '가장 작은 거리 3개 평균']

# 3. 행정동별로 가장 작은 거리 5개의 평균
top5_mean_distance = (
    df_distances.groupby('행정동_index')['거리(m)']
    .apply(lambda x: x.nsmallest(5).mean())
    .reset_index()
)
top5_mean_distance.columns = ['행정동_index', '가장 작은 거리 5개 평균']

# 결과를 하나의 데이터프레임으로 합치기
result = min_distance.merge(top3_mean_distance, on='행정동_index').merge(top5_mean_distance, on='행정동_index')
print(result)
```
