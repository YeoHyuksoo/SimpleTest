```python
import pandas as pd

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')


def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
        
```

```python
import geopy.distance
```

```python
oil_station = pd.read_csv("oil_station.csv")

oil_station
```

```python
import pandas as pd
from geopy.distance import geodesic


# 거리 계산 결과를 저장할 리스트
distances = []

# 행정동과 주유소 간 거리 계산
for i, dong_row in data.iterrows():
    if dong_row['주유소_최솟값'] > 0:
        continue
    for j, gas_row in oil_station.iterrows():
        dong_coord = (dong_row['위도'], dong_row['경도'])
        gas_coord = (gas_row['위도'], gas_row['경도'])
        distance = geodesic(dong_coord, gas_coord).meters  # 거리(m)
        distances.append([i, j, distance])

# 결과를 데이터프레임으로 변환
df_distances = pd.DataFrame(distances, columns=['행정동_index', 'oil_index', '거리(m)'])
print(df_distances)
```

```python
df_distances.to_csv("distance.csv", index=False, encoding='utf-8-sig')
```

```python
df_distances['거리(m)'].min()
```

```python

# 1. 행정동별로 주유소 간 거리의 최솟값
min_distance = df_distances.groupby('행정동_index')['거리(m)'].min().reset_index()
min_distance.columns = ['행정동_index', '최솟값']

# 2. 행정동별로 가장 작은 거리 3개의 평균
top3_mean_distance = (
    df_distances.groupby('행정동_index')['거리(m)']
    .apply(lambda x: x.nsmallest(3).mean())
    .reset_index()
)
top3_mean_distance.columns = ['행정동_index', '가장 작은 거리 3개 평균']

# 3. 행정동별로 가장 작은 거리 5개의 평균
top5_mean_distance = (
    df_distances.groupby('행정동_index')['거리(m)']
    .apply(lambda x: x.nsmallest(5).mean())
    .reset_index()
)
top5_mean_distance.columns = ['행정동_index', '가장 작은 거리 5개 평균']

# 결과를 하나의 데이터프레임으로 합치기
result = min_distance.merge(top3_mean_distance, on='행정동_index').merge(top5_mean_distance, on='행정동_index')
print(result)
```

```python
result.to_csv("access_lpg.csv", index=False, encoding='utf-8-sig')
```

```python
data_merge = data_merge.drop(['Unnamed: 0', 'EMD_CD', 'EMD_KOR_NM'], axis=1)
```

```python
gu_data = pd.read_csv("구별데이터병합.csv")

gu_data
```

```python
gu_data = gu_data.drop('Unnamed: 0', axis = 1)
```

```python
gu_data
```

```python
import pandas as pd

# '구' 컬럼을 기준으로 합치기
gu_merge = pd.merge(gu_data, data_merge, how='left', left_on = '시군구별', right_on = '구')

# 수치형 컬럼에 대해서만 평균 계산
numeric_cols = gu_merge.select_dtypes(include='number').columns
result_df = gu_merge.groupby('구', as_index=False)[numeric_cols].mean()

# 문자열 컬럼은 첫 번째 값으로 유지
string_cols = gu_merge.select_dtypes(exclude='number').columns
for col in string_cols:
    result_df[col] = gu_merge.groupby('구', as_index=False)[col].first()[col]

print(result_df)
```

```python
result_df.to_csv("final_gu_result.csv", index=False, encoding='utf-8-sig')
```

```python
result_df = result_df.drop(['지도코드', '행정코드', '위도', '경도', '시군구별', '지도_동', '시', '동'], axis = 1)
```

```python
numeric_df = result_df.select_dtypes(include='number')

# 상관분석 수행
corr_df = numeric_df.corr()
```
