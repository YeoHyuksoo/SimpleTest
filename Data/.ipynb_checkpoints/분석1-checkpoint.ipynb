```python
import folium

# 대구광역시 중심 좌표 설정
center = [35.8714, 128.6014]  # 대구광역시 중심 좌표
map = folium.Map(location=center, zoom_start=12)

# 대구광역시 데이터만 필터링 (구나 지역명에 '대구광역시'가 포함된 행만 선택)

# 대구광역시 데이터에 대해 원 추가
for i in range(len(df0)): 
    folium.Circle(
        location=[df0.iloc[i]['위도'], df0.iloc[i]['경도']],  # 각 행의 위도, 경도
        radius=150,  # 원 크기
        color='#32408c', fill_color='#32408c',
        fill_opacity=0.5
    ).add_to(map)

# 지도 출력
map
```

```python
import folium
df0 = pd.read_csv('oil_station.csv')

data_size = len(df0)

lat = df0['위도'].mean()
long = df0['경도'].mean()
map = folium.Map(location=[lat, long], zoom_start=11)

for i in range(data_size) : 
    folium.Circle(list(df0.iloc[i][['위도', '경도']]),
                  radius=20,
                  color='#32408c', fill_color='#32408c').add_to(map)
map
```

```python
import folium
from folium.plugins import MarkerCluster

lat = df0['위도'].mean()
long = df0['경도'].mean()

map_b = folium.Map((lat, long), zoom_start = 11)
mc = MarkerCluster()

for row in df0.itertuples():
    mc.add_child(folium.Marker(location = [row.위도, row.경도]))
    map_b.add_child(mc)


map_b
```
