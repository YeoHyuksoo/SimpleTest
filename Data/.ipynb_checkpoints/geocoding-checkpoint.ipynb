```python
import pandas as pd

from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')


def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
        
```

```python
# lpg_addr['ADDR'] = lpg_addr['ADDR'].apply(lambda x : re.split(r'd+', x)[0])

lpg_addr['ADDR'] = lpg_addr['ADDR'].str.extract(r'(.+\d+)', expand=False)
```

```python
lpg_addr.to_csv("lpg_addr_edit.csv", index=False, encoding='utf-8-sig')
```

```python
latitude = []
longitude =[]

for i in lpg_addr['ADDR']:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])


lpg_addr['lati'] = latitude
lpg_addr['longi'] = longitude
```
