```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv("loc_code.csv")
data
```

```python
data2 = pd.read_csv("estimate_traffic.csv")
data2
```

```python
data_merge = pd.merge(left = data , right = data2[['emd_code', 'ALL_AADT', 'PSCR_AADT', 'BUS_AADT', 'FGCR_AADT']], how = "left", left_on="code", right_on = "emd_code")

data_merge
```

```python
data_merge = data_merge.drop('emd_code', axis=1)
data_merge
```

```python
data_merge.to_csv("data_merge.csv", index=False, encoding="utf-8-sig")
```

```python
data = pd.read_csv("congestion_freq_intensity.csv")
data
```

```python
data_mapcode = pd.read_csv("loc_mapcode.csv")

data_mapcode
```

```python
import re
```

```python
data_mapcode['new_kor'] = data_mapcode['EMD_KOR_NM'].apply(lambda x: re.split(r'\d+', x)[0])

print(data_mapcode)
```

```python
data_merge = pd.merge(left = data_mapcode , right = data_merge, how = "left", left_on="new_kor", right_on = "dong")

data_merge
```

```python
data_merge.to_csv("data_merge.csv", index=False, encoding="utf-8-sig")
```
